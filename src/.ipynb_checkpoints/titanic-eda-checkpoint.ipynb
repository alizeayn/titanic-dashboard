```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv("/home/paya/Desktop/Titanic-Analysis/data/Titanic-Dataset.csv")
df.head(10)
```

```python
print(df.isnull().sum(), df.describe())
print(df.info())
print(df.columns)
```

```python
df['Age'] = df['Age'].fillna(np.round(df['Age'].mean()))
```

```python
df = pd.get_dummies(df, columns=['Embarked'])
```

```python
df = df.drop(['Ticket','Name', 'Cabin', 'PassengerId'], axis=1)
```

```python
df['Sex'] = df['Sex'].map({'male':1, 'female':0})
```

```python
df.head(50)
```

```python
print(df.corr)
```

```python
print(df['Sex'].value_counts())
print(df.groupby('Sex')['Survived'].mean())
```
